```python
#load data
df_songs = pd.read_csv('songs_with_predictions_small.csv')

# define variables
target_user = "user_5c0cba28e67a99c235ef2ba9877fef44"

# only keep the columns track_name, artist_name, and the target user
df_cleaned = df_songs[['track_name', 'artist_name', target_user]]

# rename the target user column to 'user utility rating (predicted)'
df_cleaned = df_cleaned.rename(columns={target_user: 'user utility rating (predicted)'})

# sort the cleaned dataframe by the user utility rating in descending order
df_cleaned = df_cleaned.sort_values(by='user utility rating (predicted)', ascending=False)

# reset the index of the cleaned dataframe
df_cleaned = df_cleaned.reset_index(drop=True)

# print only the first 100 rows of the cleaned dataframe with all columns aligned
with pd.option_context('display.max_rows', 100, 'display.max_columns', None, 'display.width', 1000):
    print(df_cleaned.head(100))

```
